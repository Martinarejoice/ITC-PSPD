```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid')

# Load dataset
df = pd.read_csv("paper_grades.csv")
df.head()
```

```python
# Tensile Strength by Grade
sns.barplot(data=df, x='Grade', y='Tensile Strength (MPa)', ci=None)
plt.title('Tensile Strength by Paper Grade')
plt.show()
```

```python
# Tear Resistance by Grade
sns.barplot(data=df, x='Grade', y='Tear Resistance (mN)', ci=None)
plt.title('Tear Resistance by Paper Grade')
plt.show()
```

```python
sns.boxplot(data=df, x='Grade', y='Brightness (%)')
plt.title('Brightness Distribution by Paper Grade')
plt.show()
```

```python
sns.barplot(data=df, x='Grade', y='Production Volume (tons)', estimator=sum, ci=None)
plt.title('Total Production Volume by Paper Grade (20202024)')
plt.show()
```

```python
sns.lineplot(data=df, x='Year', y='Price ($/ton)', hue='Grade', marker='o')
plt.title('Price Trends by Paper Grade (20202024)')
plt.show()
```

```python
sns.scatterplot(data=df, x='Tensile Strength (MPa)', y='Brightness (%)', hue='Grade')
plt.title('Tensile Strength vs Brightness by Grade')
plt.show()
```
